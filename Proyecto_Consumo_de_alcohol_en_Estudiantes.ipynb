#### f) Analizando a los estudiantes hombres menores a 19 años que viven en zonas urbanas ¿Notas una relación entre Pstatus y el consumo de alcohol?

```python
m = df['sex'] == 'M'
menores= df['age']<19
urbano= df['address']=='U'
filtro = df[m & menores & urbano]
filtro.groupby('Pstatus')[['Walc', 'Dalc']].mean()
```

En promedio, si influye si los padres estan separador en el consumo de alcohol de los estudiantes hombres menores de 19 años y que viven en zonas urbanas.
